```python
import fitz  # PyMuPDF for PDF text extraction
from sentence_transformers import SentenceTransformer
import chromadb

model = SentenceTransformer("all-MiniLM-L6-v2")

#Extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = "\n".join([page.get_text() for page in doc])
    return text

# Convert extracted text to embeddings
def get_embedding(text):
    return model.encode(text).tolist()

# Store in ChromaDB
def store_in_chroma(pdf_path, collection_name="pdf_embeddings"):
    # Initialize ChromaDB client
    chroma_client = chromadb.PersistentClient(path="./chroma_db")
    collection = chroma_client.get_or_create_collection(name=collection_name)

    # Extract text
    text = extract_text_from_pdf(pdf_path)

    # Generate embedding
    embedding = get_embedding(text)

    # Store in ChromaDB
    collection.add(
        ids=[pdf_path],  # Unique identifier for the PDF
        embeddings=[embedding],
        metadatas=[{"file_name": pdf_path, "text": text}]
    )

    print(f"Stored embedding for {pdf_path} in ChromaDB.")

pdf_path = "/content/BackPropDeriv.pdf"
store_in_chroma(pdf_path)
```

```python
  import chromadb
from sentence_transformers import SentenceTransformer

# Load the same embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Initialize ChromaDB client
chroma_client = chromadb.PersistentClient(path="./chroma_db")
collection = chroma_client.get_collection(name="pdf_embeddings")

# Function to get the answer
def ask_question(query):
    query_embedding = model.encode(query).tolist()

    # Perform similarity search in ChromaDB
    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=1  # Get the most relevant result
    )

    if results["documents"]:
        print("Answer:", results["metadatas"][0][0]["text"])  # Return the matched text chunk
    else:
        print("No relevant information found.")

ask_question("What is the title?")
```
